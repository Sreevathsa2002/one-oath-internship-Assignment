```python
import numpy as np
import pandas as pd
data=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\Data  (2).csv")
data
```

```python
names=list(data.iloc[:,0].values)
bright_traits=data.iloc[:,3:11].values
avg_bt=[np.sum(i)/len(i) for i in bright_traits]
print("The averages of bright traits are:")
print("\n\n")
for j in range(len(names)):
    print("{} - {}".format(names[j],avg_bt[j]))
dark_traits=data.iloc[:,11:15].values
avg_dt=[np.sum(i)/len(i) for i in dark_traits]
print("\n\n")
print("The averages of dark traits are:")
print("\n\n")
for j in range(len(names)):
    print("{} - {}".format(names[j],avg_dt[j]))
print("\n\n")

print("The overall average of bright and dark traits are:")
avg=[]
for i in range(len(bright_traits)):
    avg.append((avg_bt[i]+avg_dt[i])/2)
print("\n\n")

for j in range(len(names)):
    print("{} - {}".format(names[j],avg[j]))
```

```python
overall_dict={}
for i in range(len(names)):
    overall_dict[names[i]]=avg[i]
```

```python
#Assigning categories
A=[]
B=[]
C=[]
D=[]
for i in overall_dict:
    if overall_dict[i]>=0 and overall_dict[i]<=1:
        A.append(i)
    elif overall_dict[i]>=2 and overall_dict[i]<=3:
        B.append(i)
    elif overall_dict[i]>=4 and overall_dict[i]<=6:
        C.append(i)
    else:
        D.append(i)
C
```
